```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\deept\Desktop\AirisDATA - Project\input\top3_gross.csv")
df.info()
```

```python
country=df['country'].unique()
gross1=np.array(df.loc[df['rank']==1].gross)
gross2=np.array(df.loc[df['rank']==2].gross)
gross3=np.array(df.loc[df['rank']==3].gross)

x=np.arange(len(country))
width=0.25

fig,ax=plt.subplots()
bar1=ax.bar(x,gross1,width,label='RANK-1')
bar2=ax.bar(x+width,gross2,width,label='RANK-2')
bar3=ax.bar(x+2*width,gross3,width,label='RANK-3')

ax.set_xticks(x)
ax.set_xticklabels(country)
ax.legend()

plt.xlabel('Country\n',fontdict={'fontsize': 30, 'fontweight' : 5, 'color' : 'green'})
plt.ylabel('Gross\n',fontdict={'fontsize': 30, 'fontweight' : 5, 'color' : 'green'})
plt.title("Top 3 grossed films in each country\n",fontdict={'fontsize': 50, 'fontweight' : 5, 'color' : 'green'})

plt.xticks(rotation=90)
plt.gcf().set_size_inches(60,20)

plt.savefig(r'C:\Users\deept\Desktop\AirisDATA - Project\plot_images\top3_countryWiseGross.png')

plt.show()

```
