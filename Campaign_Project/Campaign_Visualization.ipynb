```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\deept\Desktop\AirisDATA - Project\1.Campaign_Project\campaign_write\part-00000-6c692937-4fbd-4475-94dc-6645cfe71872-c000.csv")
df.info()
```

```python
df2 = df.sort_values(by=['Total'],ascending=False)
```

```python
df3 = df2.head(n=20)
```

```python
year = df3['Each_Item']
gross = df3['Total']

plt.plot(year, gross,marker='*',markersize=4)

plt.title("Amount paid vs keyword ( highest 20 )\n", fontdict={'fontsize': 20, 'fontweight' : 5, 'color' : 'Green'})

# Labeling Axes
plt.xlabel("Each_Item", fontdict={'fontsize': 12, 'fontweight' : 5, 'color' : 'Brown'})
plt.ylabel("Total ", fontdict={'fontsize': 12, 'fontweight' : 5, 'color' : 'Brown'} )


plt.gcf().set_size_inches(15,10)
plt.xticks(rotation=90)

plt.savefig(r'C:\Users\deept\Desktop\AirisDATA - Project\1.Campaign_Project\plot_images_campaign\Amount_keyword_highest.jpeg')

plt.show()

```
